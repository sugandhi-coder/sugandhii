```python
#EV Adoption Over time  :-- Analyze the growth of population by model year


import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")



plt.figure(figsize = (12,6))

ev_adoption_by_year = ev_data['Model Year'].value_counts().sort_index()

print(ev_adoption_by_year)


sns.barplot(
    x = ev_adoption_by_year.index,
    y = ev_adoption_by_year.values,
    palette = "viridis"
)

plt.title("Electric Vehicle Adoption Over Time")
plt.xlabel("Model Year")
plt.ylabel("Number of Vehicles Registered")

plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()









```

```python
#Geographical Distribution at County Level  :-
# Selecting the top 3 Counties based on Ev Registrations and then analyze the distribuions of EVs within the cities of those Counties.



ev_county_distribution = ev_data['County'].value_counts()

print(ev_county_distribution)


top_counties = ev_county_distribution.head(3).index


print(top_counties)



#Filter tthe datasets for these top counties


top_counties_data = ev_data[ev_data['County'].isin(top_counties)]




#Analyze the distributions of EVs within the cities of these Counties

ev_city_distribution_top_counties = top_counties_data.groupby(['County','City']).size().sort_values(ascending = False).reset_index(name= 'Number of Vehicles')

print(ev_city_distribution_top_counties)



top_cities = ev_city_distribution_top_counties.head(10)

print(top_cities)


plt.figure(figsize = (12,6))

sns.barplot(
    x = 'City',
    y = 'Number of Vehicles',
    hue = 'County',
    data = top_cities,
    palette = "magma"
)

plt.title("Geographical Distribution of Electric Vehicles (Top 10 Cities)")
plt.xlabel("City")
plt.ylabel("Number of Vehicles")
plt.legend(title ='County')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()




```

```python
# Types of Electric Vehicle

ev_type_distribution = ev_data['Electric Vehicle Type'].value_counts()

print(ev_type_distribution)

plt.figure(figsize = (12,6))

sns.barplot(
    x = ev_type_distribution.index,
    y = ev_type_distribution.values,
    palette = "viridis"
)

plt.title("Types of Electric Vehicles")
plt.xlabel("Electric Vehicle Type")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()

```

```python

#Popularity of Electric Vehicle Manufacturers and models among the registered vehicles



ev_manufacturer_distribution = ev_data['Make'].value_counts()

print(ev_manufacturer_distribution.head(10))

plt.figure(figsize = (12,6))

sns.barplot(
    x = ev_manufacturer_distribution.head(10).index,
    y = ev_manufacturer_distribution.head(10).values,
    palette = "cubehelix"
)

plt.title("Popularity of Electric Vehicle Manufacturers")
plt.xlabel("Electric Vehicle Manufacturer")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()


```

```python
#Selecting the top 3 manufacturers based on the number of vehicles registered

top_3_makes = ev_manufacturer_distribution.head(3).index

print(top_3_makes)

#Filter the dataset  for those manufacturers

top_makes_data = ev_data[ev_data['Make'].isin(top_3_makes)]

print(top_makes_data)


#Analyze the popularity of EV Models within those top Manufacturers

ev_model_distribution_top_makes = top_makes_data.groupby(['Make','Model']).size().sort_values(ascending = False).reset_index(name = 'Number of Vehicles')


#Top 10 Models

top_models = ev_model_distribution_top_makes.head(10)

print(top_models)


plt.figure(figsize=(12,8))
sns.barplot(
    x = 'Model',
    y = 'Number of Vehicles',
    hue = 'Make',
    data = top_models,
    palette = "magma"
)

plt.title("Popularity of EV Models within Top Manufacturers")
plt.xlabel("Electric Vehicle Model")
plt.ylabel("Number of Vehicles")
plt.legend(title = 'Make')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show






```

```python
#Electric Range Analysis


#How far EV can travel on  a Single charge

plt.figure(figsize = (12,6))

sns.histplot(
    ev_data['Electric Range'],
    bins = 30,
    kde = True,
    color = 'royalblue'
)
plt.title("Distribution of Electric Range")
plt.xlabel("Electric Range in miles")
plt.ylabel("Number of Vehicles")
plt.axvline(ev_data['Electric Range'].mean(), color = 'red', linestyle = '--', label = 'Mean')
plt.tight_layout()
plt.show()







```

```python
#Calculating the Average electric range by model year

average_range_by_year = ev_data.groupby('Model Year')['Electric Range'].mean().reset_index()

print(average_range_by_year)


plt.figure(figsize = (12,6))

sns.lineplot(
    x = 'Model Year',
    y = 'Electric Range',
    data = average_range_by_year,
    marker = 'o',
    color = 'green'
)

plt.title("Average Electric Range by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Average Electric Range")
plt.grid(True)
plt.show()


```

```python
ev_registration_counts = ev_data['Model Year'].value_counts().sort_index()

print(ev_registration_counts)
```

```python
#Forecasting :- Cmpound Annual Growth Rate

from scipy.optimize import curve_fit
import numpy as np


#Filter the datasets to inclue the year with complete data

filtered_years = ev_registration_counts[ev_registration_counts.index <= 2023]

print(filtered_years)

#Definen a function for exponential Growth to fit the data

def exponential_growth(t, a, b):
    return a * np.exp(b * t)


#Prepare the data for curve fitting

x_data = filtered_years.index - filtered_years.index.min()
y_data = filtered_years.values

print(x_data)


#Fit the data to exponential groiwth function

params, covariance = curve_fit(exponential_growth, x_data, y_data)

#Use the filetered function to forecaset the number of EVs in 2024 and next five years

forecast_years = np.arange(2024, 2024+6) - filtered_years.index.min()
print(forecast_years)

forecast_values = exponential_growth(forecast_years, *params)

print(forecast_values)


#Create a dictionary to diplay the forecasted values

forcasted_evs = dict(
    zip(
        forecast_years + filtered_years.index.min(),
        forecast_values
    )
)

print(forcasted_evs)








```

```python
#prepare the data for plotting

years = np.arange(filtered_years.index.min(), 2029 + 1)

actual_years = filtered_years.index

forecast_years_full = np.arange(2024, 2029 + 1)


actual_values = filtered_years.values
forecast_values_full = [forcasted_evs[year] for year in forecast_years_full]

plt.figure(figsize=(12,8))

plt.plot(
    actual_years,
    actual_values,
    marker = 'o',
    linestyle = '-',
    color = 'blue',
    label = 'Actual'
)

plt.plot(
    forecast_years_full,
    forecast_values_full,
    marker = 'o',
    linestyle = '--',
    color = 'red',
    label = 'Forecast'
)
plt.title("Current and Estimated EV Market")
plt.xlabel("Year")
plt.ylabel("Number of Vehicles")
plt.legend()
plt.grid(True)
plt.show()






```
